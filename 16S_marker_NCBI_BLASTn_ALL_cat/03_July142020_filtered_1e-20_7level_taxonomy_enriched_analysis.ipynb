```python
plt.rcParams["figure.figsize"] = (30,8)
colors = sns.color_palette('pastel')[0:11]
df['Class-phyceae'].value_counts().plot.pie(labeldistance=1.15, 
                                     wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, 
                                     colors = colors,
                                     autopct='%1.1f%%',
                                     ylabel='')
```

```python
plt.rcParams["figure.figsize"] = (30,8)
colors = sns.color_palette('pastel')[0:11]
df['Class-phyceae'].value_counts()[1:101].plot.pie(labeldistance=1.15, 
                                     wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, 
                                     colors = colors,
                                     autopct='%1.1f%%',
                                     ylabel='')
```

```python
plt.figure(figsize=(16, 10))
colors = sns.color_palette('Paired')[0:55]
df['Class-phyceae'].value_counts()[1:101].plot.bar(color = colors, )
```
