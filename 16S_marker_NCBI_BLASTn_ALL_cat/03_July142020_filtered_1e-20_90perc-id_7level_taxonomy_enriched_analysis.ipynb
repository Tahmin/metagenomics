### Visualization | Community Composition/Taxonomic distribution for July142020

```python
df = pd.read_csv('07_July142020_filtered_1e-20_90perc-id_subj_to_7level_taxonomy_enriched_analysis.csv')
```

```python
df.head(5)
```

```python
df.shape
```

```python
df[['Phylum-phyta','Class-phyceae']].isna().sum()
```

```python
plt.rcParams["figure.figsize"] = (30,8)
colors = sns.color_palette('pastel')[0:11]
df['Class-phyceae'].value_counts().plot.pie(labeldistance=1.15, 
                                     wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, 
                                     colors = colors,
                                     autopct='%1.1f%%',
                                     ylabel='')
```

```python
plt.rcParams["figure.figsize"] = (30,8)
colors = sns.color_palette('pastel')[0:11]
df['Class-phyceae'].value_counts()[1:10].plot.pie(labeldistance=1.15, 
                                     wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, 
                                     colors = colors,
                                     autopct='%1.1f%%',
                                     ylabel='')
```

```python
plt.figure(figsize=(16, 10))
colors = sns.color_palette('Paired')[0:55]
df['Class-phyceae'].value_counts()[1:101].plot.bar(color = colors, )
```

```python
plt.rcParams["figure.figsize"] = (30,8)
colors = sns.color_palette('pastel')[0:11]
df['Phylum-phyta'].value_counts().plot.pie(labeldistance=1.15, 
                                     wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, 
                                     colors = colors,
                                     autopct='%1.1f%%',
                                     ylabel='')
```

```python
plt.rcParams["figure.figsize"] = (30,8)
colors = sns.color_palette('pastel')[0:11]
df['Phylum-phyta'].value_counts()[1:5].plot.pie(labeldistance=1.15, 
                                     wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, 
                                     colors = colors,
                                     autopct='%1.1f%%',
                                     ylabel='')
```
